```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Plot styling
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.family'] = 'DejaVu Sans'

print('✅ Libraries imported successfully')
```

```python
df = pd.read_csv('amazon.csv')
print(f'Dataset shape: {df.shape[0]} rows × {df.shape[1]} columns')
df.head()
```

```python
print('='*50)
print('COLUMN NAMES & DATA TYPES')
print('='*50)
print(df.dtypes)
print()
print('='*50)
print('BASIC STATISTICS')
print('='*50)
df.describe(include='all')
```

```python
# --- Missing Values ---
print('MISSING VALUES PER COLUMN:')
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct.round(2)})
print(missing_df[missing_df['Missing Count'] > 0])

# Drop rows where key fields are missing
df.dropna(subset=['rating', 'discounted_price', 'actual_price'], inplace=True)
print(f'\nAfter dropping rows with missing key fields: {df.shape[0]} rows remain')
```

```python
# --- Fix Data Types ---

# Remove currency symbols and commas, convert to float
def clean_price(col):
    return col.astype(str).str.replace('[₹,]', '', regex=True).str.strip().astype(float)

df['discounted_price'] = clean_price(df['discounted_price'])
df['actual_price']     = clean_price(df['actual_price'])

# Remove % sign from discount_percentage
df['discount_percentage'] = (
    df['discount_percentage'].astype(str)
    .str.replace('%', '', regex=False)
    .str.strip()
    .astype(float)
)

# Convert rating to numeric (some may have text)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Remove commas from rating_count and convert
df['rating_count'] = (
    df['rating_count'].astype(str)
    .str.replace(',', '', regex=False)
    .str.strip()
)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Extract top-level category (before first |)
df['main_category'] = df['category'].astype(str).str.split('|').str[0].str.strip()

print('✅ Data types cleaned.')
print(df[['discounted_price','actual_price','discount_percentage','rating','rating_count']].dtypes)
```

```python
# --- Duplicate Check ---
dupes = df.duplicated().sum()
print(f'Duplicate rows: {dupes}')
df.drop_duplicates(inplace=True)
print(f'Final dataset size: {df.shape[0]} rows × {df.shape[1]} columns')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
axes[0].hist(df['rating'].dropna(), bins=20, color='steelblue', edgecolor='white')
axes[0].set_title('Distribution of Product Ratings', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Rating')
axes[0].set_ylabel('Count')
axes[0].axvline(df['rating'].mean(), color='red', linestyle='--', label=f"Mean: {df['rating'].mean():.2f}")
axes[0].legend()

# Boxplot
axes[1].boxplot(df['rating'].dropna(), patch_artist=True,
                boxprops=dict(facecolor='lightblue', color='navy'),
                medianprops=dict(color='red', linewidth=2))
axes[1].set_title('Boxplot of Ratings', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Rating')
axes[1].set_xticks([])

plt.tight_layout()
plt.savefig('fig_rating_distribution.png', bbox_inches='tight')
plt.show()

print(f"\nRating Stats:")
print(df['rating'].describe().round(2))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Actual Price
axes[0].hist(df['actual_price'].dropna(), bins=40, color='coral', edgecolor='white')
axes[0].set_title('Distribution of Actual Price (₹)', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Price (₹)')
axes[0].set_ylabel('Count')
axes[0].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'₹{x:,.0f}'))

# Discount %
axes[1].hist(df['discount_percentage'].dropna(), bins=30, color='mediumpurple', edgecolor='white')
axes[1].set_title('Distribution of Discount %', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Discount (%)')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('fig_price_distribution.png', bbox_inches='tight')
plt.show()

print('\nActual Price Stats:')
print(df['actual_price'].describe().round(2))
print('\nDiscount % Stats:')
print(df['discount_percentage'].describe().round(2))
```

```python
top_categories = df['main_category'].value_counts().head(10)

plt.figure(figsize=(12, 6))
bars = plt.barh(top_categories.index[::-1], top_categories.values[::-1],
                color=sns.color_palette('Blues_d', len(top_categories)))
plt.title('Top 10 Product Categories by Count', fontsize=14, fontweight='bold')
plt.xlabel('Number of Products')

for bar, val in zip(bars, top_categories.values[::-1]):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
             str(val), va='center', fontsize=10)

plt.tight_layout()
plt.savefig('fig_category_distribution.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
plt.scatter(df['discount_percentage'], df['rating'],
            alpha=0.4, color='teal', edgecolors='none', s=25)

# Trend line — use paired dropna to keep lengths equal
clean = df[['discount_percentage', 'rating']].dropna()
z = np.polyfit(clean['discount_percentage'], clean['rating'], 1)
p = np.poly1d(z)
x_line = np.linspace(clean['discount_percentage'].min(), clean['discount_percentage'].max(), 100)
plt.plot(x_line, p(x_line), color='red', linewidth=2, linestyle='--', label='Trend Line')

plt.title('Discount % vs Product Rating', fontsize=14, fontweight='bold')
plt.xlabel('Discount (%)')
plt.ylabel('Rating')
plt.legend()
plt.tight_layout()
plt.savefig('fig_discount_vs_rating.png', bbox_inches='tight')
plt.show()

corr = df[['discount_percentage', 'rating']].corr().iloc[0, 1]
print(f'Pearson Correlation (Discount % vs Rating): {corr:.4f}')
```

```python
cat_rating = (
    df.groupby('main_category')['rating']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure(figsize=(12, 6))
bars = sns.barplot(data=cat_rating, x='rating', y='main_category',
                   palette='RdYlGn', orient='h')
plt.title('Average Rating by Product Category (Top 10)', fontsize=14, fontweight='bold')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.xlim(0, 5)

for i, (val, _) in enumerate(zip(cat_rating['rating'], cat_rating['main_category'])):
    plt.text(val + 0.03, i, f'{val:.2f}', va='center', fontsize=10)

plt.tight_layout()
plt.savefig('fig_rating_by_category.png', bbox_inches='tight')
plt.show()
```

```python
sample = df.sample(min(300, len(df)), random_state=42)

plt.figure(figsize=(9, 5))
plt.scatter(sample['actual_price'], sample['discounted_price'],
            alpha=0.5, color='darkorange', s=30, edgecolors='none')
# Perfect discount line (0% off)
max_val = max(sample['actual_price'].max(), sample['discounted_price'].max())
plt.plot([0, max_val], [0, max_val], 'r--', linewidth=1.5, label='No Discount (y=x)')

plt.title('Actual Price vs Discounted Price (₹)', fontsize=14, fontweight='bold')
plt.xlabel('Actual Price (₹)')
plt.ylabel('Discounted Price (₹)')
plt.legend()
plt.tight_layout()
plt.savefig('fig_price_comparison.png', bbox_inches='tight')
plt.show()
```

```python
num_cols = ['actual_price', 'discounted_price', 'discount_percentage', 'rating', 'rating_count']
corr_matrix = df[num_cols].corr()

plt.figure(figsize=(9, 6))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, linewidths=0.5, vmin=-1, vmax=1,
            annot_kws={'size': 11})
plt.title('Correlation Heatmap — Numeric Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('fig_correlation_heatmap.png', bbox_inches='tight')
plt.show()
```

```python
def detect_outliers_iqr(series, label):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = series[(series < lower) | (series > upper)]
    print(f'{label}: Q1={Q1:.2f}, Q3={Q3:.2f}, IQR={IQR:.2f}')
    print(f'  Bounds: [{lower:.2f}, {upper:.2f}]')
    print(f'  Outliers: {len(outliers)} ({len(outliers)/len(series)*100:.1f}%)')
    print()
    return lower, upper

print('=== IQR OUTLIER DETECTION ===')
for col in ['actual_price', 'discount_percentage', 'rating', 'rating_count']:
    detect_outliers_iqr(df[col].dropna(), col)
```
